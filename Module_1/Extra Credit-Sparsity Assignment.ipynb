```python
#Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
import tarfile
```

```python
#Loading the Data from Github Repository
username = "harisx91"
path = "Recommender_System_DAV_6300/master/Module_1/jester-data-1.csv"
df = pd.read_csv(f'https://raw.githubusercontent.com/{username}/{path}')
```

```python
df.head()
```

```python
#Looking for null values
df.isnull().sum()
```

```python
print(df.shape)
```

```python
##The values "99.0" corresponds to "null"= "not rated". Replaces those values to 0
df_01= df.replace(99.0, 0)
df_01.head()
```

```python
##Transforming the data from wide to long 
df_02 = df_01.copy()
df_02.index = np.arange(1, len(df_01) + 1)
df_02
```

```python
#Extract ratings columns
ratings = df_02.iloc[:, 1:]
ratings
```

```python
df_03 = ratings.stack().reset_index()
df_03.columns=['UserId', 'jokeID', 'rating']

df_03
```

```python
##Sparsity of the matrix: 
##Sparsity Matrix = Number of Empty cells/Total number of cells

(df_03['rating']==0).sum() / 2498200
```

There are about 27.75% of the cells that are empty

```python
#Fit StandardScaler into ratings
from sklearn.preprocessing import StandardScaler

df_03[['rating']] = StandardScaler().fit_transform(df_03[['rating']])
df_03
```
