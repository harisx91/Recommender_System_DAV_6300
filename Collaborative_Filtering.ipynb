### Data Prep

```python
movies_average_rating=df_survey.groupby('movie')['rating'].mean().sort_values(ascending=False).reset_index().rename(columns={'rating':'Average Rating'})
movies_average_rating.head()
```

Here I took the average rating for each movie

```python
movies_rating_count=df_survey.groupby('movie')['rating'].count().sort_values(ascending=True).reset_index().rename(columns={'rating':'Rating Count'}) #ascending=False
movies_rating_count_avg=movies_rating_count.merge(movies_average_rating, on='movie')
movies_rating_count_avg.head()
```

I added another rating count column so we can see how many ratings each movie has

```python
rating_with_RatingCount = df_survey.merge(movies_rating_count, left_on = 'movie', right_on = 'movie', how = 'left')
rating_with_RatingCount.head()
```

Merged the two datasets for further data extraction

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(rating_with_RatingCount['Rating Count'].describe())
```

Added function so we have three decimal points for our statistical inference

```python
popularity_threshold = 3
popular_movies= rating_with_RatingCount[rating_with_RatingCount['Rating Count']>=popularity_threshold]
popular_movies.head()
#popular_movies.shape
```

```python
import os
movie_features_df=popular_movies.pivot_table(index='movie',columns='UserID',values='rating').fillna(0)
movie_features_df.head(10)
```

### KNN Model

```python
from scipy.sparse import csr_matrix
movie_features_df_matrix = csr_matrix(movie_features_df.values)
```

```python
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
movie_features_df.shape
```

```python
query_index = np.random.choice(movie_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(movie_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
query_index=2
```

```python
movie_features_df.head()
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(movie_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, movie_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```

```python
from lightfm import LightFM
from lightfm.datasets import fetch_movielens
from lightfm.evaluation import precision_at_k
from lightfm.cross_validation import random_train_test_split
import scipy.sparse as sp
```

```python
ratings = np.array(popular_movies['rating'])
users = np.array(popular_movies['UserID'])
items = np.array(popular_movies['movie'])
```

```python
from sklearn.preprocessing import LabelEncoder

# heavy lifting encoders
user_encoder = LabelEncoder()
item_encoder = LabelEncoder()
```

```python
# preparation for the csr matrix
u = user_encoder.fit_transform(users)
i = item_encoder.fit_transform(items)
lu = len(np.unique(u))
li = len(np.unique(i))
```

```python
matrix = sp.coo_matrix((ratings, (u, i)), shape=(lu, li))
```

```python
matrix.todense()
```

```python
item_encoder.classes_[:7]
```

```python
user_encoder.classes_[:7]
```

```python
##item_encoder.inverse_transform([7])
```

```python
from lightfm.cross_validation import random_train_test_split

train, test = random_train_test_split(matrix, test_percentage=0.2)
```

```python
model = LightFM()
model.fit(train)
```

```python
from lightfm.evaluation import precision_at_k

precision_at_k(model, test, k=10).mean()
```

```python
user = 1
user_id = int(user_encoder.transform([user])[0])
```

```python
user_encoder.inverse_transform([user_id])
```

```python
all_movie_ids = np.array(list(range(len(item_encoder.classes_))),dtype=np.int32)
all_movie_ids.dtype
```

```python
preds = model.predict(user_id, all_movie_ids)
```

```python
all_movies = pd.DataFrame(
    zip(item_encoder.classes_, preds),
    columns=['Movie', 'Prediction']
).sort_values(
    'Prediction', ascending=False
)

all_movies.head(10)
```
